**Logarithemic Tranformation**

```python
df=pd.read_csv('titanic.csv',usecols=['Age','Fare','Survived'])
```

```python
df.isnull().sum()
```

```python
df['Fare']=df['Fare'].fillna(df['Fare'].median())
df['Age']=df['Age'].fillna(df['Age'].median())
```

```python
## to check if your feature is gaussian / normal distributed we use QQplot
```

```python
##import scipy library for QQplot
import scipy.stats as stat
import pylab
```

```python
def plot_data(df,feature):
  plt.figure(figsize=(10,6))
  plt.subplot(1,2,1)
  df[feature].hist()
  plt.subplot(1,2,2)
  stat.probplot(df[feature],dist='norm',plot=pylab)
  plt.show()
```

```python
plot_data(df,'Age')
```

#if the points are following the straight line then that feature is normally distributed else not

```python
plot_data(df,'Fare')
```

**Logarithmic Transformation**

```python
import numpy as np
df['Age_log']=np.log(df['Age'])
plot_data(df,'Age')
```

not following straight line hence move to next transformation

**Reciprocal Tranformation**

```python
df['Age_r']=1/df.Age
plot_data(df,'Age_r')
```

this as well is not following straight line so we move to 3rd transformation

**Square Root Tranformation**

```python
 df['Age_sq']=df['Age']**(1/2)
 plot_data(df,'Age_sq')
```

Again its not following the straight line

**Exponential Transformation**

```python
df['Age_exp']=df['Age']*(1/1.2)
plot_data(df,'Age_exp')
```

Pretty much most of the following the line

**Box Cox Transformation**

Formula: T(Y) = (Y exp(Lambda)-1)/Lambda

Lambda: is the paramter values of boxcox function

```python
df['Age_bc'],parameters=stat.boxcox(df['Age'])
plot_data(df,'Age_bc')
```
